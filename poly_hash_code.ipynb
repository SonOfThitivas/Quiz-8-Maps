```python
def poly_hash_code(x, a):
    result = 0
    for i in range(len(x)):                     # loops all characters
        result += ord(x[i]) * a**(len(x)-1-i)   # polynomial function
    return result
```

# Case Example #

```python
print(poly_hash_code("Hello", 1))
print(poly_hash_code("Hello", 2))
print(poly_hash_code("Hello", 33))
```

```python
print(poly_hash_code("temp01", 1))
print(poly_hash_code("temp10", 1))
```

```python
print(poly_hash_code("temp01", 37))
print(poly_hash_code("temp10", 37))
```

```python
print(poly_hash_code("stop", 1))
print(poly_hash_code("tops", 1))
print(poly_hash_code("pots", 1))
print(poly_hash_code("spot", 1))
```

```python
print(poly_hash_code("stop", 37))
print(poly_hash_code("tops", 37))
print(poly_hash_code("pots", 37))
print(poly_hash_code("spot", 37))
```
